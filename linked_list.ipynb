```python
#implement node
class Node(object):
    """This class creates a node for a linked list
    """
    def _init_(self,data=None,next_node=None):
        """constructor which initializes the node object
        input: data to be stored in node
               pointer to the next node
        return: none 
        
        """
        self.data=data
        self.next_node=next_node
        
    def get_data(self):
        """the get data method helps to obtained data stored at a node
        input:none
        return: data
        """
        return self.data
    
    def get_next(self):
        """function for obtaining the next node in a linked list from the current node
        return: next node in linked list
        """
        return self.next_node
    
    def set_next(self,new_next):
        """A function for setting a node to a differet variable 
           return: None
        """
        self.next_node=new_next
    
    
```

```python
class LinkedList(object):
    """Implementation of the linked list class
    """
    def _init_(self,head=None):
        """Constructor which intializes the head of the linked list
        """
    self.head=head
    
    def insert(self,data):
        """Method to insert data into a node on the linked list
        input:data
        returns: None
        """
        new_node=Node(data)
        new_node.set_next(self.head)
        self.head=new_node
        
    def size(self):
        """This method finds the size of a linked list
        input: None
        return:size of linked list (number of nodes)
        """
        current=self.head
        count=0
        while current:
            count+=1
            current=current.get_next()
        return count
    
    def search(self,data):
        """ The search method loops through the linked list to find data(node) of interest
        input:data to be found
        return: Node where data is stored
        """
        current=self.head
        found=False
        while current and found is False:
            if current.get_data()==data:
                found=True
            else:
                current=current.get_next()
        if current is None:
            raise ValueError("Data not found")
        return current
    def delete(self,data):
        """The delete method removes 
        """
        current=self.head
        previous=None
        found=False
        while current and found is False:
            if current.get_data()==data:
                found=True
            else:
                previous=current
                current=current.get_next()
        if current is None:
            raise ValueError("Data not in list")
        if previous is None:
            self.head = current.get_next()
        else:
            previous.set_next(current.get_next())
            
        
                
                
            
            
```
